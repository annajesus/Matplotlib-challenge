## Analysis

* The correlation of 0.84 between mouse weight and average tumor volume showed that as the mouse weight increases, the average tumor volume does also. 
* The data revealed that Ramicane was the most effective drug and following was Capomulin. They both yielded the lowest mean, median, variance, and standard deviation.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress


# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)


```

```python
# Combine the data into a single dataset
combined_data_results_df = pd.merge(mouse_metadata, study_results, how="left", on="Mouse ID")

# Display the data table for preview
combined_data_results_df
```

```python
# Checking the number of mice. (ans 249)
len(combined_data_results_df['Mouse ID'].unique())
```

```python
duplicate_mice_id = combined_data_results_df.loc[combined_data_results_df.duplicated(subset=['Mouse ID', 'Timepoint']), 'Mouse ID'].unique()
duplicate_mice_id
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mouse_id_data = combined_data_results_df.loc[combined_data_results_df["Mouse ID"] == "g989"]
duplicate_mouse_id_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = combined_data_results_df[combined_data_results_df['Mouse ID'].isin(duplicate_mice_id)==False]
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
mice_clean_df = clean_df["Mouse ID"].nunique()
mice_clean_df
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.


tumor_vol_mean = round(clean_df.groupby('Drug Regimen').mean()["Tumor Volume (mm3)"],2)
tumor_vol_median = round(clean_df.groupby('Drug Regimen').median()["Tumor Volume (mm3)"],2)
tumor_vol_variance = round(clean_df.groupby('Drug Regimen').var()["Tumor Volume (mm3)"],2)
tumor_vol_std = round(clean_df.groupby('Drug Regimen').std()["Tumor Volume (mm3)"],2)
tumor_vol_sem = round(clean_df.groupby('Drug Regimen').sem()["Tumor Volume (mm3)"],2)

summary_df = pd.DataFrame({"Mean Tumor Volume":tumor_vol_mean,"Median Tumor Volume":tumor_vol_median,"Tumor Volume Variance":tumor_vol_variance,"Tumor Volume Std. Dev.":tumor_vol_std,"Tumor Volume Std. Err.":tumor_vol_sem})

summary_df


```

```python
# Using the aggregation method, produce the same summary statistics in a single line.
summary_stats_agg = round(clean_df.groupby(['Drug Regimen'])[['Tumor Volume (mm3)']].agg(['mean', 'median', 'var', 'std','sem']),2)
summary_stats_agg
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
last_timepoint=clean_df.groupby(['Mouse ID'])['Timepoint'].max()
last_tp = last_timepoint.to_frame()
last_tp
```

```python
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
tumor_vol_last_tp_df = last_tp.merge(clean_df, on=(["Mouse ID","Timepoint"]))
tumor_vol_last_tp_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
```

```python
# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data_list = []
```

```python
# Calculate the IQR and quantitatively determine if there are any potential outliers.
# Locate the rows which contain mice on each drug and get the tumor volumes
for treatment in treatments:
    drug_regimen = tumor_vol_last_tp_df.loc[tumor_vol_last_tp_df["Drug Regimen"] == treatment]
    new_tumor_vol = drug_regimen['Tumor Volume (mm3)']
#     print(drug_regimen)

# #append tumor data to new list 
    tumor_vol_data_list.append(new_tumor_vol)
#     print(tumor_vol_data_list)
    quartile = new_tumor_vol.quantile([.25,.5,.75])
    lowerq = quartile[0.25]
    upperq = quartile[0.75]
    iqr = upperq-lowerq
    
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    

# print(f"{treatments} potential outliers are lower quartile: {lowerq}, upper quartile: {upperq}, and interquartile: {iqr}.")
# print(f"Ramicane's potential outliers: lower quartile: {lowerq}, upper quartile: {upperq}, and interquartile: {iqr}.")
# print(f"Infubinol's potential outliers: lower quartile: {lowerq}, upper quartile: {upperq}, and interquartile: {iqr}.")
# print(f"Ceftamin's potential outliers: lower quartile: {lowerq}, upper quartile: {upperq}, and interquartile: {iqr}.")
    
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.

fig1, ax1 = plt.subplots(figsize=(10,5))
ax1.boxplot(tumor_vol_data_list, labels=treatments)
ax1.set_title('Tumor Volume by Drug Regimen', fontsize=20)
ax1.set_ylabel('Final Tumor Volume (mm3)', fontsize=12)
ax1.set_xlabel('Drug Regimen',fontsize = 12)

plt.show()
```

## Correlation and Regression

```python
# # Calculate the correlation coefficient and linear regression model 
# # for mouse weight and average tumor volume for the Capomulin regimen
# # y = mx + b


# Use same scatter plot from before to start
plt.scatter(avg_tumor_vol, avg_weight) 
plt.title('Tumor Vol vs Mouse Weight')
plt.xlabel('Avg Weight (g)')
plt.ylabel('Avg Tumor Volume (mm3)')

plt.show()
```

```python
# Add the linear regression equation and line to plot
x_values = vol_vs_weight_df['Weight (g)']
y_values = vol_vs_weight_df['Tumor Volume (mm3)']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.scatter(x_values, y_values)
plt.plot(x_values, regress_values, 'r-')
plt.annotate(line_eq,(6,12),fontsize=12,color="red")
plt.xlabel('Weight (g)')
plt.ylabel('Avg Tumor Volume (mm3)')

# correlation equation {round(st.pearsonr(arg1, arg2)[0],2)}

print(f"The correlation between mouse weight and the average tumor volume is {round(st.pearsonr(x_values, y_values)[0],2)}")
print(f"The r-squared is: {rvalue**2}")

plt.show()
```
